```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# === Données ===
data = {
    'Projet': ['Tech Task', 'MFAaaS', 'NLG', 'Dataviz', 'IRS', 'RUN'],
    'Prévu': [16, 37, 13, 12, 13, 20],
    'Réalisé': [9, 30, 8, 4, 5, 18],
    'Mercato': [7, 7, 5, 7, 0, 2]
}
df = pd.DataFrame(data)

# === Préparation ===
y = np.arange(len(df))
fig, ax = plt.subplots(figsize=(6, 3))
bar_height = 0.25

# --- Barres ---
#ax.barh(y - bar_height/2, df['Prévu'], height=bar_height, color='#CFE9E6', label='Prévu')
#ax.barh(y + bar_height/2, df['Réalisé'], height=bar_height, color='#00796B', label='Réalisé')
ax.barh(y - bar_height/2, df['Prévu'], height=bar_height, color='#00A2A3', label='Prévu')
ax.barh(y + bar_height/2, df['Réalisé'], height=bar_height, color='#8ACBBB', label='Réalisé')

# --- Mercato entrant ---
for i, (mercato, realise, proj_y) in enumerate(zip(df['Mercato'], df['Réalisé'], y)):
    if mercato > 0:
        ax.barh(proj_y + bar_height/2, mercato, left=realise, height=bar_height, color='#8B0000',
                label='Mercato entrant' if i == 0 else "")

# --- Lignes verticales en pointillés (tous les 10 points, uniquement sur l'axe X) ---
for x in range(10, 46, 10):
    ax.axvline(x=x, color='gray', linestyle='--', linewidth=0.7, alpha=0.3, ymin=0, ymax=1)

# --- Configuration visuelle ---
ax.set_yticks(y)
ax.set_yticklabels(df['Projet'], fontsize=10)
ax.set_xlim(0, 45)
ax.set_xlabel("")
ax.tick_params(axis='x', length=0)
ax.tick_params(axis='y', length=0)
ax.legend(fontsize=7, loc='upper center', ncol=3, bbox_to_anchor=(0.5, -0.15))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.grid(False)

plt.tight_layout()
plt.show()
```
